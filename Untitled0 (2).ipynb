```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_iris

data = load_iris()

X = data.data
y = data.target

print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
X = np.array([20, 22, 25, 30, 35]).reshape(-1,1)  # Temperature
y = np.array([100, 120, 150, 200, 250])          # Ice cream sales

# Scatter plot
plt.scatter(X, y, color='orange')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales ($)')
plt.title('Temperature vs Ice Cream Sales')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Step 1: Create the model
model = LinearRegression()

# Step 2: Train the model on the data
model.fit(X, y)
```

```python
import matplotlib.pyplot as plt

# Predict sales for all X values
y_pred = model.predict(X)

# Plot the points and the learned line
plt.scatter(X, y, color='orange')  # original points
plt.plot(X, y_pred, color='blue')  # line learned by model
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales ($)')
plt.title('ML Learned Line')
plt.show()
```

```python
# Predict sales for a new temperature
new_temp = np.array([[28]]) #28 temperature ho
predicted_sales = model.predict(new_temp)
print("Predicted ice cream sales at 28°C:", predicted_sales[0])
```

```python
print("Slope (per °C):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error

# Predict for training data
y_pred_train = model.predict(X)

# Calculate MSE
mse = mean_squared_error(y, y_pred_train)
print("Mean Squared Error:", mse)

# mse 9.7x10^-28 aaxa, almost 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([15, 18, 20, 22, 25, 28, 30, 32, 35]).reshape(-1,1)

# Sales with some random noise
np.random.seed(42)  # for reproducibility
B = np.array([80, 95, 100, 120, 150, 180, 200, 220, 250]) + np.random.randint(-10, 10, size=9)

# Plot
plt.scatter(X, y, color='orange')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales ($)')
plt.title('Temperature vs Ice Cream Sales (Noisy Data)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
A = np.array([15, 18, 20, 22, 25, 28, 30, 32, 35]).reshape(-1,1)

# Sales with some random noise
np.random.seed(42)  # for reproducibility
B = np.array([80, 95, 100, 120, 150, 180, 200, 220, 250]) + np.random.randint(-10, 10, size=9)

model = LinearRegression()

model.fit(A,B)

B_pred = model.predict(A)

plt.scatter(A, B, color='orange')  # original points
plt.plot(A, B_pred, color='blue')  # line learned by model
```

```python
from sklearn.metrics import mean_squared_error

# MSE for noisy data
mse_noisy = mean_squared_error(B, B_pred)
print("Mean Squared Error (noisy data):", mse_noisy)

#mse large xa cuz of the noise
```

```python
# Predict sales for a new temperature, e.g., 27°C
new_temp = np.array([[27]])
predicted_sales = model.predict(new_temp)
print("Predicted ice cream sales at 27°C:", predicted_sales[0])
```
